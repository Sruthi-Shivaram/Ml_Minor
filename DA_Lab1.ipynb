```python
import matplotlib.pyplot as plt
correlation=data.corr()
correlation
plt.figure(figsize=(12,10))

sns.heatmap(correlation,annot=True,cmap=plt.cm.Reds)
plt.show
```

The above plot shows the correlation matrix of the dataset. The correlation values lie between -1 and 1 with 1 showing highest correlation or similarity whereas -1 shows inverse relation between the two attributes.

```python
correlation_mat = data.corr()
corr_pairs = correlation_mat.unstack()

sorted_pairs = corr_pairs.sort_values(kind="quicksort")
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]
print(strong_pairs)
```

The above snippet is used to extract the pair of attributes which have values greater than 0.5. This is to check which attributes show strong correlation which can then be used for further analysis.

```python

sns.distplot(data['fatal_accidents_1985_to_1999'])
```

distplot is used for getting the distribution plot of the data attribute
The above plot shows distribution of the fatal accidents in the year 1985 to 1999. According to the plot, most airlines have 0 to 5 fatal accidents in the period with a few airlines having as high as 10 to 20 fatal accidents between 1985 to 1999.


```python
sns.relplot(x='fatal_accidents_1985_to_1999',y='incidents_1985_to_1999',hue='airline',data=data)
```

relplot is used to plot relational plot between two variables.
The above plot shows the relation between fatal accidents and incidents in the period 1985 to 1999. It shows around 2 to 4 fatal accidents have occured for most airines which have around 10 to 20 incidents. However, Aeroflot airline shows very high amount fatal incidents from the incidents that have occured. This airline data could be considered as an outlier.






```python
sns.pairplot(data[['fatal_accidents_2000_to_2014','incidents_2000_to_2014']])
```

The above plot shows a pairplot of the number of fatal incidents that occured from the total airline accidents in the period 2000 to 2014. The plot shows most airlines have around 0 to 2 fatal accidents with a very few airlines having high number of fatal accidents.

```python
print(data['incidents_2000_to_2014'].quantile(0.5))
print(data['incidents_2000_to_2014'].quantile(0.75))
print(data['incidents_2000_to_2014'].quantile(0.95))
```

The function quantile is used to find the frequency of values lying intervals of 25% of the total amound of data.
The above statistical parameter shows on an average there are 3 incidents for most airlines with few airlines having 5 or 11 incidents in the period 2000 to 2014. 

```python
sns.catplot(x='avail_seat_km_per_week',kind='box',data=data)
```


The above catplot function is used plot categorical plots. It shows the box plot of 'avail_seat_km_per_week' attribute of the dataset which depicts the available seat kilometers flown every week. Around 10 to 20 crore km is flown every week by most airlines with few airlines having very high values of 60 to 70 crore kms per week.

```python
data['%_fatalities_from_1985_to_1999']=(data['fatal_accidents_1985_to_1999']*100)/data['incidents_1985_to_1999']
data['%_fatalities_from_2000_to_2014']=(data['fatal_accidents_2000_to_2014']*100)/data['incidents_2000_to_2014']
data.head()
```

The above function is used to calculate the percentage of fatalities from the total incidents that occur for each airline in the period 1985 to 2014. The function calculates the percentage and assigns it to a new attribute.

```python
data.isnull().sum()
```

The modified dataset has some null values which need to be cleaned for analysis. 

```python
data=data.dropna()
data.isnull().sum()
```

The null values are dropped using the dropna() function and the data set is now free from null values.

```python
data.nunique()
```

We need to get the unique values to find out the top ten airlines that have reported the most fatalities in the period 1985 to 2014

```python
new=data.sort_values(by="%_fatalities_from_1985_to_1999",ascending=False,kind="mergesort")
new=new.iloc[0:10]
fig = plt.figure(figsize = (40, 10))
# creating the bar plot
plt.bar(new['airline'], new['%_fatalities_from_1985_to_1999'], color ='blue',
        width = 0.4)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 

plt.xlabel("Airlines")
plt.ylabel("Perecentage fatalities between 1985 and 1999")
plt.title("Top 10 Fatality percentage for each airline in the years 1985 to 1999")
plt.show()
```

.bar() plot is used to plot a bar graph using two variables
The above plot shows the top 10 airlines with most percentage of fatalities in the period 1985 to 1999 .The highest percentage of fatality is reported by EI AI airlines of Isreal.

```python
new1=data.sort_values(by="%_fatalities_from_2000_to_2014",ascending=False,kind="mergesort")
new1=new1.iloc[0:10]
fig = plt.figure(figsize = (30, 5))
# creating the bar plot
plt.bar(new1['airline'], new1['%_fatalities_from_2000_to_2014'], color ='orange',
        width = 0.4)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 
plt.xlabel("Airlines")
plt.ylabel("Perecentage fatalities between 2000 and 2014")
plt.title(" Top 10 Fatality percentage for each airline in the years 2000 to 2014")
plt.show()
```

The above plot shows the top 10 airlines with most percentage of fatalities in the period 2000 to 2014 .The highest percentage of fatality is reported by Kenya Airways in the above period.

```python
data['total_fatal_accidents']=data['fatal_accidents_1985_to_1999']+data['fatal_accidents_2000_to_2014']
data.nunique()
```

The above attribute shows the total fatal accidents of each airline for further analysis.

```python
new2=data.sort_values(by="total_fatal_accidents",ascending=False,kind="mergesort")
new2=new2.iloc[0:10]
fig = plt.figure(figsize = (30, 5))
# creating the bar plot
plt.bar(new2['airline'], new2['total_fatal_accidents'], color ='magenta',
        width = 0.5)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 
plt.xlabel("Airlines")
plt.ylabel("Fatal accidents")
plt.title("Top 10 Fatal incidents for each airline from 1985 to 2014")
plt.show()
```

The above bar graph shows the total fatal accidents for each airline in the period 1985 to 2014. The highest number of fatal accidents occured for Aeroflot with 14 fatal accidents. Initially in the relational plot it showed high values hence was considered an outlier.The above graph confirms the high number of accidents occuring for the airline.

```python
sns.relplot(x="fatalities_2000_to_2014", y="incidents_2000_to_2014", hue="airline", size="avail_seat_km_per_week",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=data)
```

The fatalities can be high due to avaliable seat km per week. This shows that Aeroflot airlines has high number of fatalities by considering other attributes.

```python
new4=data.sort_values(by="total_fatal_accidents",ascending=True,kind="mergesort")
new4=new4.iloc[0:10]
fig = plt.figure(figsize = (30, 5))
# creating the bar plot
plt.bar(new4['airline'], new4['total_fatal_accidents'], color ='magenta',
        width = 0.5)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 
plt.xlabel("Airlines")
plt.ylabel("Fatal accidents")
plt.title("Top 10 least fatal incidents for each airline from 1985 to 2014")
plt.show()
```

The above plot shows the top 10 airlines with least number of fatal accidents in the period 1985 to 2014. There are 6 airlines having 0 fatal incidents.

```python
airlines = data["airline"]
total_number_of_incidents = data["incidents_2000_to_2014"]+data["incidents_1985_to_1999"]
fig = plt.figure(figsize =(20, 15))
plt.title("Number of incidents happened from 1985 to 2014")
plt.rcParams.update({'font.size': 10})
plt.pie(total_number_of_incidents)
plt.legend(title = "AirLines", labels = airlines)
plt.show()

```

.pie() is used to plot the pie chart for a given variable.
The above pie chart shows the maximum number of incidents that occured in the period 1985 to 2014. The chart shows maximum number of incidents have occured for Turkish Airlines.
