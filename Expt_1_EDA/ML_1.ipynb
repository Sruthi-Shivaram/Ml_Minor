# Bivariate Analysis

```python
import matplotlib.pyplot as plt
correlation=data.corr()
correlation
plt.figure(figsize=(12,10))

sns.heatmap(correlation,annot=True,cmap=plt.cm.Reds)
plt.show
```

The above plot shows correlation values of different attributes.

```python
correlation_mat = data.corr()
corr_pairs = correlation_mat.unstack()

sorted_pairs = corr_pairs.sort_values(kind="quicksort")
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]
print(strong_pairs)
```

The above snippet lists the attributes showing strong correlation with each other. The above shows girls in the age group below 12 and 12 to 16 have a strong correlation in the number of crimes being committed.  

```python
sns.relplot(x='Cases Reported against Juveniles - (Col. 3)',y='Total - Total - (Col. 15)',hue='Crime Head (Col. 2)',data=data)
```

The above plot verifies the fact that the total cases reported against juvenile are extremely high with the highest cases being of a Total Cognizable IPC crimes.

```python
sns.pairplot(data[['Below 12 Years - Girls - (Col. 5)','16 Years & Above and below 18 Years - Girls - (Col. 11)']])
```

The above plot is a pairplot. It shows strong positive correlation between the two attributes ,'Girls of the age 16 and above' and 'girls below 12 yrs age group'. Both of committed around 250 crimes in the year 2020.

```python
print(data['Total - Total - (Col. 15)'].quantile(0.5))
print(data['Total - Total - (Col. 15)'].quantile(0.75))
print(data['Total - Total - (Col. 15)'].quantile(0.95))
```

On an average the crimes comitted by juvenile in each type is around 74 with a few crimes having as high as 7299 crimes in the year 2020.

```python
new=data.sort_values(by="Total - Total - (Col. 15)",ascending=False,kind="mergesort")
new=new.iloc[0:10]
fig = plt.figure(figsize = (40, 10))
# creating the bar plot
plt.bar(new['Crime Head (Col. 2)'], new['Total - Total - (Col. 15)'], color ='blue',
        width = 0.4)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 

plt.xlabel("Crimes")
plt.ylabel("Total cases")
plt.title("Top 10 crime casesin juvenile")
plt.show()
```

The above bar plot shows the top 10 cases of crimes committed by juveniles.The highest being Total Cognizable IPC crimes followd by Offences affecting human body.

```python
new=data.sort_values(by="Total - Girls - (Col. 14)",ascending=False,kind="mergesort")
new=new.iloc[0:10]
fig = plt.figure(figsize = (40, 10))
# creating the bar plot
plt.bar(new['Crime Head (Col. 2)'], new['Total - Girls - (Col. 14)'], color ='pink',
        width = 0.4)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 

plt.xlabel("Crimes")
plt.ylabel("Total cases by girls")
plt.title("Top 10 crime cases in juvenile")
plt.show()
```

The top 10 crimes committed by girls is Total Cognizable IPC being the highest followed by offences against human body. The maximum crime committed by girls is 350. Girls show a high number of Hurt but less number of Theft crimes.

```python
new=data.sort_values(by="Total - Boys - (Col. 13)",ascending=False,kind="mergesort")
new=new.iloc[0:10]
fig = plt.figure(figsize = (40, 10))
# creating the bar plot
plt.bar(new['Crime Head (Col. 2)'], new['Total - Boys - (Col. 13)'], color ='red',
        width = 0.4)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 

plt.xlabel("Crimes")
plt.ylabel("Total cases by boys")
plt.title("Top 10 crime cases in juvenile by boys")
plt.show()
```

The above plot shows the top 10 crimes committed by boys in the year 2020. Boys have committed very high number of crimes compared to girls. The highest crime committed is of Cognizable IPC offences with the number rising upto 30000. They have a significant amount of high theft cases compared to girls.

```python
new=data.sort_values(by="Below 12 Years - Total - (Col. 6)",ascending=False,kind="mergesort")
new=new.iloc[0:10]
fig = plt.figure(figsize = (40, 10))
# creating the bar plot
plt.bar(new['Crime Head (Col. 2)'], new['Below 12 Years - Total - (Col. 6)'], color ='green',
        width = 0.4)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 

plt.xlabel("Crimes")
plt.ylabel("Total cases by boys")
plt.title("Top 10 crime cases in juvenile below 12 years")
plt.show()
```

The above plot shows the crime comitted by juveniles below 12 yrs. The crime committed shows a high value for IPC crimes but also shows high values for Rape and murder.

```python
Crime_head = data["Crime Head (Col. 2)"]
total_number_of_incidents_12_16 = data["12 Years & Above and below 16 Years - Total - (Col. 9)"]
fig = plt.figure(figsize =(20, 15))
plt.title("Crimes by girls in the age 12 to 16")
plt.rcParams.update({'font.size': 10})
plt.pie(total_number_of_incidents_12_16)
plt.legend(title = "Crime Head (Col. 2)", labels = Crime_head)
plt.show()
```

In case of girls in the age group 12 to 16 there are a high number of murder cases along with total cognizable IPC crimes.

```python
Crime_head = data["Crime Head (Col. 2)"]
total_number_of_incidents_12_below = data["16 Years & Above and below 18 Years - Total - (Col. 12)"]
fig = plt.figure(figsize =(20, 15))
plt.title("Crimes by juveniles under the age 12")
plt.rcParams.update({'font.size': 10})
plt.pie(total_number_of_incidents_12_16)
plt.legend(title = "Crime Head (Col. 2)", labels = Crime_head)
plt.show()
```

The above plot shows the crime cases of juveniles in the year 16 years to 18 years. They show a high crime rate for IPC crimes like all age groups.
