```python
print(max(data['Total - Total - (Col. 15)']))
sns.distplot(data['Total - Total - (Col. 15)'])
```

```python
sns.catplot(x='Cases Reported against Juveniles - (Col. 3)',kind='box',data=data)
```

```python
import matplotlib.pyplot as plt
correlation=data.corr()
correlation
plt.figure(figsize=(12,10))

sns.heatmap(correlation,annot=True,cmap=plt.cm.Reds)
plt.show
```

```python
correlation_mat = data.corr()
corr_pairs = correlation_mat.unstack()

sorted_pairs = corr_pairs.sort_values(kind="quicksort")
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]
print(strong_pairs)
```

```python
sns.relplot(x='Cases Reported against Juveniles - (Col. 3)',y='Total - Total - (Col. 15)',hue='Crime Head (Col. 2)',data=data)
```

```python
sns.pairplot(data[['Below 12 Years - Girls - (Col. 5)','16 Years & Above and below 18 Years - Girls - (Col. 11)']])
```

```python
print(data['Total - Total - (Col. 15)'].quantile(0.5))
print(data['Total - Total - (Col. 15)'].quantile(0.75))
print(data['Total - Total - (Col. 15)'].quantile(0.95))
```

```python
new=data.sort_values(by="Total - Total - (Col. 15)",ascending=False,kind="mergesort")
new=new.iloc[0:10]
fig = plt.figure(figsize = (40, 10))
# creating the bar plot
plt.bar(new['Crime Head (Col. 2)'], new['Total - Total - (Col. 15)'], color ='blue',
        width = 0.4)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 

plt.xlabel("Crimes")
plt.ylabel("Total cases")
plt.title("Top 10 crime casesin juvenile")
plt.show()
```

```python
new=data.sort_values(by="Total - Girls - (Col. 14)",ascending=False,kind="mergesort")
new=new.iloc[0:10]
fig = plt.figure(figsize = (40, 10))
# creating the bar plot
plt.bar(new['Crime Head (Col. 2)'], new['Total - Girls - (Col. 14)'], color ='pink',
        width = 0.4)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 

plt.xlabel("Crimes")
plt.ylabel("Total cases by girls")
plt.title("Top 10 crime casesin juvenile")
plt.show()
```

```python
new=data.sort_values(by="Total - Boys - (Col. 13)",ascending=False,kind="mergesort")
new=new.iloc[0:10]
fig = plt.figure(figsize = (40, 10))
# creating the bar plot
plt.bar(new['Crime Head (Col. 2)'], new['Total - Boys - (Col. 13)'], color ='red',
        width = 0.4)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 

plt.xlabel("Crimes")
plt.ylabel("Total cases by boys")
plt.title("Top 10 crime cases in juvenile by boys")
plt.show()
```

```python
Crime_head = data["Crime Head (Col. 2)"]
total_number_of_incidents_12_16 = data["12 Years & Above and below 16 Years - Total - (Col. 9)"]
fig = plt.figure(figsize =(20, 15))
plt.title("Crimes by girls in the age 12 to 16")
plt.rcParams.update({'font.size': 10})
plt.pie(total_number_of_incidents_12_16)
plt.legend(title = "Crime Head (Col. 2)", labels = Crime_head)
plt.show()
```
